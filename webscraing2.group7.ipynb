```python
#Write a Python program to extract all the URLs from the webpage python.org that are nested within <li> tags from
import requests
from bs4 import BeautifulSoup

# URL to scrape
url = "https://www.python.org"

# Fetch the webpage
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find all <li> tags and extract URLs within <a> tags
links = soup.find_all('li')

print("URLs nested within <li> tags:\n")
for li in links:
    a_tag = li.find('a')
    if a_tag and 'href' in a_tag.attrs:
        link = a_tag['href']
        print(link)
```

```python
#Write a Python program to extract all the text from a given web page.
import requests
from bs4 import BeautifulSoup

def extract_text_from_webpage(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the text from the parsed HTML
        text = soup.get_text(separator=' ', strip=True)

        return text
    else:
        return f"Error: Unable to retrieve the page. Status code: {response.status_code}"

# URL for Python official website
url = 'https://www.python.org'
webpage_text = extract_text_from_webpage(url)

# Print the first 1000 characters of the extracted text for preview
print(webpage_text[:1000])
```

```python
#Write a Python program to retrieve the HTML code of the title, its text, and the HTML code of its parent.


import requests
from bs4 import BeautifulSoup

def retrieve_title_info(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the title tag
        title_tag = soup.find('title')

        if title_tag:
            # Get the title text
            title_text = title_tag.get_text()

            # Get the HTML code of the title tag
            title_html = str(title_tag)

            # Get the parent of the title tag
            title_parent_html = str(title_tag.find_parent())

            return {
                'title_text': title_text,
                'title_html': title_html,
                'title_parent_html': title_parent_html
            }
        else:
            return "No title tag found on this webpage."
    else:
        return f"Error: Unable to retrieve the page. Status code: {response.status_code}"

# Example usage
url = 'https://www.python.org'  # Replace with the desired URL
title_info = retrieve_title_info(url)

# Print the result
print("Title Text:", title_info.get('title_text'))
print("Title HTML:", title_info.get('title_html'))
print("Parent HTML of Title:", title_info.get('title_parent_html'))
```

```python
#Write a Python program to create a Beautiful Soup parse tree into a nicely formatted Unicode string, with a separate line for each HTML/XML tag and string.
import requests
from bs4 import BeautifulSoup

def get_pretty_html(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Get the prettified HTML in Unicode format
        pretty_html = soup.prettify()

        return pretty_html
    else:
        return f"Error: Unable to retrieve the page. Status code: {response.status_code}"

# Example usage
url = 'https://www.python.org'  # Replace with the desired URL
pretty_html = get_pretty_html(url)

# Print the formatted output
print(pretty_html)
```

```python
#Write a Python program to remove the contents of a tag in a given html document.
import requests
from bs4 import BeautifulSoup

def remove_tag_contents(url, tag_to_clear):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all instances of the specified tag
        tags = soup.find_all(tag_to_clear)

        # Loop through all found tags and clear their contents
        for tag in tags:
            tag.clear()  # This will remove the contents of the tag

        return soup.prettify()  # Return the modified HTML document
    else:
        return f"Error: Unable to retrieve the page. Status code: {response.status_code}"

# Example usage
url = 'https://www.python.org'  # Replace with the desired URL
tag_to_clear = 'div'  # Replace with the tag you want to clear
modified_html = remove_tag_contents(url, tag_to_clear)

# Print the modified HTML
print(modified_html)
```

```python
#Write a Python program to wrap an element in the specified tag and create the new wrapper.
import requests
from bs4 import BeautifulSoup

def wrap_element_in_tag(url, tag_to_wrap, wrapper_tag):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the first occurrence of the tag you want to wrap
        element_to_wrap = soup.find(tag_to_wrap)

        if element_to_wrap:
            # Create the new wrapper tag
            wrapper = soup.new_tag(wrapper_tag)

            # Wrap the element in the new tag
            element_to_wrap.wrap(wrapper)

            return soup.prettify()  # Return the modified HTML document
        else:
            return f"Error: No element found with the tag '{tag_to_wrap}'"
    else:
        return f"Error: Unable to retrieve the page. Status code: {response.status_code}"

# Example usage
url = 'https://www.python.org'  # Replace with the desired URL
tag_to_wrap = 'p'  # Tag you want to wrap (e.g., 'p' for paragraph)
wrapper_tag = 'div'  # The new wrapper tag (e.g., 'div')

modified_html = wrap_element_in_tag(url, tag_to_wrap, wrapper_tag)

# Print the modified HTML
print(modified_html)
```

```python
#Write a Python program to remove a tag from a given tree of html document and destroy it and its contents.
import requests
from bs4 import BeautifulSoup

def remove_tag_and_contents(url, tag_to_remove):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all instances of the specified tag
        tags = soup.find_all(tag_to_remove)

        # Loop through all found tags and decompose (remove) them and their contents
        for tag in tags:
            tag.decompose()

        return soup.prettify()  # Return the modified HTML document
    else:
        return f"Error: Unable to retrieve the page. Status code: {response.status_code}"

# Example usage
url = 'https://www.python.org'  # Replace with the desired URL
tag_to_remove = 'div'  # Replace with the tag you want to remove
modified_html = remove_tag_and_contents(url, tag_to_remove)

# Print the modified HTML
print(modified_html)
```
